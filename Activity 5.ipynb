```python
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
internet = pd.read_csv('Dataset//Internet-Demographic.csv')
```

```python
internet.info()
```

```python
internet['Country Name'] = internet['Country Name'].astype('category')
internet['Country Code'] = internet['Country Code'].astype('category')
internet['Income Group'] = internet['Income Group'].astype('category')
```

```python
internet.dtypes
```

```python
internet.isnull().sum()
```

<h1>1. DISTRIBUTION OF BIRTH RATE</h1>

```python
plt.hist(internet["Birth rate"], bins=5, color='blue', edgecolor='black')
plt.xlabel('Birth Rate')
plt.ylabel('Frequency')
plt.title('Distribution of Birth Rates')
plt.grid(True)
plt.show()
```

<h1>2.Top 5 Countries with Highest Internet Usage</h1>

```python
top_5 = internet.nlargest(5, "Internet users")  # Get top 5 countries

plt.bar(top_5["Country Name"], top_5["Internet users"], color='green')
plt.xlabel('Country')
plt.ylabel('Internet Users (%)')
plt.title('Top 5 Countries by Internet Usage')
plt.show()
```

<h1>3. Count of Countries by Income Group</h1>

```python
sns.countplot(x=internet["Income Group"], palette='cool')
plt.xlabel('Income Group')
plt.ylabel('Number of Countries')
plt.title('Countries by Income Group')
plt.show()
```

<h1>4. Box Plot of Internet Usage
</h1>

```python
sns.boxplot(y=internet["Internet users"], color='yellow')
plt.ylabel('Internet Users (%)')
plt.title('Box Plot of Internet Usage')
plt.show()
```

<h1>5. Countries with no Internet Access</h1>

```python
no_internet = internet[internet["Internet users"] == 0][["Country Name", "Country Code"]]
print("Countries with No Internet Access:")
print(no_internet)
```

<h1>6. Average Birth Rate</h1>

```python
avg_birth_rate = internet["Birth rate"].mean()
print(f'Average Birth Rate: {avg_birth_rate:.2f}')
```

<h1>7. Scatter Plot: Birth Rate VS. Internet Usage</h1>

```python
plt.scatter(internet["Birth rate"], internet["Internet users"], color='green', alpha=0.7)
plt.xlabel('Birth Rate')
plt.ylabel('Internet Users (%)')
plt.title('Birth Rate vs Internet Usage')
plt.grid(True)
plt.show()
```

<h1>8. Countries with Birth Rates Above 40</h1>

```python
high_birth_rate = internet[internet['Birth rate'] > 40][['Country Name', 'Birth rate']]
(high_birth_rate)
```

<h1>9. Pie Chart of Income Groups</h1>

```python
internet["Income Group"].value_counts().plot.pie(autopct='%1.1f%%', colors=['pink', 'yellow', 'green', 'orange'])
plt.title('Income Group Distribution')
plt.ylabel('')
plt.show()
```

<h1>10. Correlation Between Birth Rate and Internet Usage</h1>

```python
correlation = internet["Birth rate"].corr(internet["Internet users"])
print(f'Correlation between Birth Rate and Internet Users: {correlation:.2f}')
```
