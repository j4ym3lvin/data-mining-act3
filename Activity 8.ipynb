```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings 
import folium
from folium.plugins import HeatMap
from IPython.display import display
warnings.filterwarnings('ignore')
```

```python


# Load the dataset with the correct encoding
df = pd.read_csv('Dataset\\Chicago_Crimes.csv', encoding='ISO-8859-1')

```

```python
category_cols = ['Case Number', 'Block', 'IUCR', 'Primary Type',
                 'Description', 'Location Description', 'FBI Code', 'Location']

df[category_cols] = df[category_cols].astype('category')
```

```python
df.dtypes
```

```python
df.isnull().sum()
```

```python
df["Primary Type"].unique()
```

<H1>THEFT incidents from 2015 onwards in Chicago</H1>

```python
fil1 = df[df['Primary Type'] == 'THEFT']
fil2 = fil1[fil1['Year'] >= 2015]
fil3 = fil2.dropna(subset=['Latitude', 'Longitude'])

# Group by location to count incidents
crime_grouped = fil3.groupby(['Longitude', 'Latitude']).size().reset_index(name='incident_count')
heat_data = crime_grouped[['Latitude', 'Longitude', 'incident_count']].values.tolist()

# Create the map centered on Chicago
mymap = folium.Map(location=[41.8781, -87.6298], zoom_start=11)

# Add heatmap
HeatMap(heat_data).add_to(mymap)

# Show map
mymap
```
